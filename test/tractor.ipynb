```python
import h5py

fpath = '/Users/kisuklee/Data_local/datasets/pinky/ground_truth/stitched/crop_x1024_y1024_z100/img.h5'

with h5py.File(fpath, 'r') as f:
    data = f['main'][...].astype('float32')
    data /= 255.0
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def random_patch(w=256):
    z = np.random.randint(0, data.shape[-3])
    y = np.random.randint(0, data.shape[-2] - w)
    x = np.random.randint(0, data.shape[-1] - w)
    return np.copy(data[z,y:y+256,x:x+256])
```

```python
import numpy as np
import augmentor
from augmentor.perturb import *
```

```python
def generate_stencil(m=20, w1=70, w2=120, b1=5, b2=10, b3=3, thresh=0.5):
    margin = (w2-w1)//2

    fig = plt.figure(figsize=(10,10))

    grad = np.zeros((256,w2)).astype('float32')
    grad[:,m:-m] = 1
    Blur(b1)(grad)
    plt.subplot(1,9,1)
    plt.imshow(grad, cmap='gray')

    stencil = np.random.rand(256,w2).astype('float32')
    plt.subplot(1,9,2)
    plt.imshow(stencil, cmap='gray')

    Blur(b2)(stencil)
    plt.subplot(1,9,3)
    plt.imshow(stencil, cmap='gray')

    stencil = (stencil > thresh).astype(stencil.dtype)
    plt.subplot(1,9,4)
    plt.imshow(stencil, cmap='gray')

    stencil *= grad
    plt.subplot(1,9,5)
    plt.imshow(stencil, cmap='gray')

    Blur(b3)(stencil)
    plt.subplot(1,9,6)
    plt.imshow(stencil, cmap='gray')

    plt.show()
    
    return stencil
```

```python
sten1 = generate_stencil()
sten2 = generate_stencil(m=30, thresh=0.503)

img = random_patch()

fig = plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')

img = img.astype(sten1.dtype)
fill = np.zeros((256,w2), dtype=sten1.dtype)

fill = 0
img[:,70:190] = img[:,70:190] * (1-sten2) +  fill * sten2

fill = 1
img[:,70:190] = img[:,70:190] * (1-sten1) +  fill * sten1

plt.subplot(1,2,2)
plt.imshow(img, cmap='gray')

plt.show()
```
